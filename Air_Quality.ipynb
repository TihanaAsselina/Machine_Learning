```python
# Data Cleaning
df.isnull().sum()
```

```python
dfs = df.drop(['ID NO','NMHC(GT)'],axis = 1)
```

```python
# splite x as independent and y as predicated
X = dfs.iloc[:,:-1]
Y = dfs.iloc[:,-1]
```

```python
X
```

```python
Y
```

```python
# TRAIN TEST SPLIT
#split for row

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(X,Y,test_size=0.25,random_state=10)
```

```python
y_test
```

```python
#Train the simple linear regression
from sklearn.linear_model import LinearRegression
air_qut =LinearRegression()
```

```python
# Providing the training / model building
air_qut.fit(x_train,y_train)
```

```python
# Predict the result
y_predict = air_qut.predict(x_test)
```

```python
y_predict
```

```python
# EVALUATION
y_test #error test
```

```python
sns.regplot(x=y_test, y=y_predict, ci=None, color="red")
plt.plot(x_train, air_qut.predict(x_train), color="pink")
plt.xlabel("CO(GT)	PT08.S1(CO)	C6H6(GT)	PT08.S2(NMHC)	NOx(GT)	PT08.S3(NOx)	NO2(GT)	PT08.S4(NO2)	PT08.S5(O3)	T	RH")
plt.ylabel("AH")
```

```python
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
```

```python
MAE = mean_absolute_error(y_test,y_predict)
```

```python
print(MAE)
```

```python
MSE = mean_squared_error(y_test,y_predict)
```

```python
print(MSE)
```

```python
RSME = np.sqrt(MSE)
```

```python
print(RSME)
```

```python
LRSME = np.log(RSME)
```

```python
print(LRSME)
```
