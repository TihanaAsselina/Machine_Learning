```python
# LABEL ENCODING:
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
df['Outlook'] = le.fit_transform(df['Outlook'])
df['Temperature'] = le.fit_transform(df['Temperature'])
df['Humidity'] = le.fit_transform(df['Humidity'])
df['Wind'] = le.fit_transform(df['Wind'])
df['Play_Badminton'] = le.fit_transform(df['Play_Badminton'])

```

```python
X = df.iloc[:,:4]
```

```python
X
```

```python
Y = df.iloc[:, 4].values
```

```python
Y
```

```python
from sklearn.model_selection import train_test_split, cross_val_score
x_train,x_test,y_train,y_test = train_test_split(X,Y,test_size=0.25,random_state=10)
```

```python
x_train
```

```python
y_train
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```python
# NAIVE BAYES USING GAUSSIANNB METHOD:
from sklearn.naive_bayes import GaussianNB
BD = GaussianNB()
BD.fit(x_train, y_train)
```

```python
y_pred = BD.predict(x_test)
y_pred
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
cm
```

```python
# NAIVE BAYES USING BERNOULLINB METHOD:
from sklearn.naive_bayes import BernoulliNB
BD1 = BernoulliNB()
BD1.fit(x_train, y_train)
```

```python
y_pre = BD1.predict(x_test)
y_pre
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pre)
cm
```
