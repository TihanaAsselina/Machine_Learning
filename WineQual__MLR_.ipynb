```python
# Dealing with missing values:
# fixed acidity = (10/6497)*100 = 0.15
# volatile acidity = (8/6497)*100 = 0.12
# citric acid = (3/6497)*100 = 0.04
# residual sugar = (2/6497)*100 = 0.03
# chlorides = (2/6497)*100 = 0.03
# pH = (9/6497)*100 = 0.13
# sulphates = (4/6497)*100 = 0.06
```

```python
dfs=df.dropna()
```

```python
dfs
```

```python
# LABEL ENCODING:
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
dfs['type']=le.fit_transform(dfs['type'])
```

```python
# splite x as independent and y as predicated
x = np.array(dfs.iloc[:, :-1])
y = np.array(dfs['quality'])
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split, cross_val_score
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.25,random_state=100)
```

```python
y_test
```

```python
# Providing the training / model building:
from sklearn.linear_model import LinearRegression
wine_qty =LinearRegression()
wine_qty.fit(x_train,y_train)
```

```python
# Predict the result
y_predict = wine_qty.predict(x_test)
y_predict
```

```python
# EVALUATION
y_test #error test
```

```python
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
```

```python
MAE = mean_absolute_error(y_test,y_predict)
print(MAE)
```

```python
MSE = mean_squared_error(y_test,y_predict)
print(MSE)
```

```python
R2 = r2_score(y_test,y_predict) # ACCURACY MEASURE
print(R2)
```
