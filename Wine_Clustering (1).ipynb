### **K - MEANS CLUSTERING :**

```python
# Elbow Method
from sklearn.cluster import KMeans
wcss = [] #within cluster sum of square
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', n_init=10, max_iter=300, random_state=0)
    kmeans.fit(x)
    wcss.append(kmeans.inertia_)
```

```python
plt.plot(range(1,11), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of Clusters')
plt.ylabel('wcss')
plt.show()
```

```python
#fitting kmeans to dataset
kmeans = KMeans(n_clusters=4, init='k-means++',n_init=10, max_iter=30, random_state=0)
Y_kmeans = kmeans.fit_predict(x)
```

```python
Y_kmeans
```

```python
plt.scatter(x[Y_kmeans==0, 0], x[Y_kmeans==0, 1], s=50, c='red', label= 'Cluster 1')
plt.scatter(x[Y_kmeans==1, 0], x[Y_kmeans==1, 1], s=50, c='cyan', label= 'Cluster 2')
plt.scatter(x[Y_kmeans==2, 0], x[Y_kmeans==2, 1], s=50, c='green', label= 'Cluster 3')
plt.scatter(x[Y_kmeans==3, 0], x[Y_kmeans==3, 1], s=50, c='blue', label= 'Cluster 4')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=200, c='yellow', label='Centroids' )
plt.title('Wine')
plt.xlabel('Total_Phenols')
plt.ylabel('Flavanoids')
plt.legend()
plt.show()
```

```python
from sklearn import metrics
print (metrics.silhouette_score(x,kmeans.labels_,metric='euclidean'))
```

### **HIERARCHICAL CLUSTERING :**

```python
# Dendogram Method
import scipy.cluster.hierarchy as sch
d_gram = sch.dendrogram(sch.linkage(x, method='ward'))
plt.title('Dendogram Method')
plt.xlabel('Alcohol')
plt.ylabel('Euclidean Distances')
plt.savefig('img_1.png', dpi=500)
plt.show()
```

```python
# Fitting Hierarchical Cluster Analysis to dataset
from sklearn.cluster import AgglomerativeClustering
hc = AgglomerativeClustering(n_clusters=4, affinity="euclidean", linkage='ward')
Y_hc = hc.fit_predict(x)
```

```python
Y_hc
```

```python
# Visualising the clusters
plt.scatter(x[Y_hc==0, 0], x[Y_hc==0, 1], s=50, c='red', label= 'Cluster 1')
plt.scatter(x[Y_hc==1, 0], x[Y_hc==1, 1], s=50, c='cyan', label= 'Cluster 2')
plt.scatter(x[Y_hc==2, 0], x[Y_hc==2, 1], s=50, c='green', label= 'Cluster 3')
plt.scatter(x[Y_hc==3, 0], x[Y_hc==3, 1], s=50, c='blue', label= 'Cluster 4')
plt.title('Wine')
plt.xlabel('Total_Phenols')
plt.ylabel('Flavanoids')
plt.legend()
plt.show()
```
