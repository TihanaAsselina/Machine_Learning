```python
#2.write a program that plots marks obtained by two students in four subjects using bar graph

Shreedevi = (69,50,45,90,49)
Saratha = (78,90,67,90,58)
index = np.array([6,4,8,3,9])
width = 0.25
plt.bar(index,Shreedevi,width,label = 'a', color='pink')
plt.bar(index+width,Saratha, width , label ="b", color='purple')
```

```python
kathir = np.array([70,80,97,70,67])
plt.bar(index,Shreedevi, width = 0.5, label = 'Shreedevi_marks', color = 'black',bottom = Saratha+kathir)
plt.bar(index, Saratha, width = 0.5, label = ' Saratha_marks',color = 'gray',bottom = kathir)
plt.bar(index, kathir, width = 0.5, label = ' Saratha_marks',color = 'pink')
```

```python
X = ['Group A','Group B','Group C','Group D']
Ygirls = [10,20,20,40]
Zboys = [20,30,25,30]
X_axis = np.arange(len(X))
plt.bar(X_axis - 0.2, Ygirls, 0.4, label = 'Girls',color='brown')
plt.bar(X_axis + 0.2, Zboys, 0.4, label = 'Boys',color='green')
plt.xticks(X_axis, X)
plt.xlabel("Groups")
plt.ylabel("Number of Students")
plt.title("Number of Students in each group")
plt.legend()
plt.show()
```
